```python
from Bio.PDB import PDBParser, PDBIO
from nglview import show_structure_file
import tempfile

# DNA Structure Generator
def generate_dna_structure():
    structure = """
ATOM      1  P     A A   1      11.000  14.600   6.500  1.00  0.00           P  
ATOM      2  O1P   A A   1      10.500  13.500   7.500  1.00  0.00           O  
ATOM      3  O2P   A A   1      10.500  15.800   5.500  1.00  0.00           O  
ATOM      4  O5'   A A   1      13.000  14.600   6.500  1.00  0.00           O  
ATOM      5  C1'   A A   1      14.000  15.600   7.000  1.00  0.00           C  
ATOM      6  N1    A A   1      15.500  15.800   7.500  1.00  0.00           N  
ATOM      7  C2    A A   1      16.500  14.800   7.000  1.00  0.00           C  
ATOM      8  N3    A A   1      16.000  13.600   6.500  1.00  0.00           N  
ATOM      9  C4    A A   1      14.500  13.500   6.000  1.00  0.00           C  
TER      10      A A   1                                                      
END
"""
    return structure

# Save the structure to a temporary .pdb file
structure_content = generate_dna_structure()
with tempfile.NamedTemporaryFile(delete=False, suffix=".pdb") as temp_file:
    temp_file.write(structure_content.encode())
    temp_pdb_file = temp_file.name

# Visualize the DNA structure using nglview
view = show_structure_file(temp_pdb_file)

# Customize the visualization
view.add_representation('cartoon', selection='all', color='blue')
view.add_representation('spacefill', selection='all', color='green')
view.center()
view
```
