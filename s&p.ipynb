```python
import matplotlib.pyplot as plt

# Create subplots, 2 rows, sharing the x-axis
fig, ax = plt.subplots(2, sharex=True, figsize=(10, 8))

# Select the first 20 rows of data
data = sp500_data.iloc[:50]

# Plot daily frequency data with no fill on the first subplot
data.asfreq('D').plot(ax=ax[0], marker='o')
ax[0].set_title('Original Frequency (No Fill)')

# Plot daily frequency data with back-fill and forward-fill on the second subplot

data.asfreq('D', method='bfill').plot(ax=ax[1], style='--o')
data.asfreq('D', method='bfill').plot(ax=ax[1], style='-o')
data.asfreq('D', method='bfill').plot(ax=ax[1], style='--')
data.asfreq('D', method='ffill').plot(ax=ax[1], style=':')

ax[1].legend(["back-fill", "forward-fill"])
ax[1].set_title('Filled Missing Data')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
### SHOW GAPS ONLY ###

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Select the first 50 rows of data (or your own dataset)
data = sp500_data.iloc[:50]

# Resample the data to daily frequency to expose gaps
resampled_data = data.asfreq('D')

# Find where the gaps (NaN values) are
gaps = resampled_data[resampled_data.isna().any(axis=1)]

# Create a new figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through the gaps and plot lines for each gap
for i in range(1, len(gaps)):
    prev_valid = resampled_data.loc[:gaps.index[i-1]].last_valid_index()
    next_valid = resampled_data.loc[gaps.index[i]:].first_valid_index()

    if prev_valid is not None and next_valid is not None:
        # Plot a line showing the break (gap) between valid points
        ax.plot([prev_valid, next_valid], [resampled_data.loc[prev_valid, 'Close_SPY'], 
                                           resampled_data.loc[next_valid, 'Close_SPY']], 
                linestyle='--', color='red', label='Gap' if i == 1 else '')

# Add labels and title
ax.set_title('Lines Indicating Breaks (Gaps) in the Data')
ax.set_xlabel('Date')
ax.set_ylabel('SPY Close Price')

# Add a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ensure the data is set to a daily frequency and forward-fill missing values
sp500_data = sp500_data.asfreq('D', method='pad')

# Calculate the Return on Investment (ROI) over 30 days using the 'Adj Close_SPY' column
ROI_30d = 100 * (sp500_data['Adj Close_SPY'].shift(-30) - sp500_data['Adj Close_SPY']) / sp500_data['Adj Close_SPY']

# Drop NaN values that may result from the shift operation
ROI_30d = ROI_30d.dropna()

# Check if ROI_30d is empty
if ROI_30d.empty:
    print("No valid data for ROI calculation over 30 days. Ensure the dataset covers at least 30 days.")
else:
    # Plot the ROI
    ROI_30d.plot()
    plt.ylabel('% Return on Investment after 30 days')
    plt.title('S&P 500 ROI After 30 Days Based on Adj Close Price')
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Ensure the data is set to a daily frequency and forward-fill missing values
sp500_data = sp500_data.asfreq('D', method='pad')

# Calculate the Return on Investment (ROI) over 30 days using the 'Adj Close_SPY' column
ROI_365d = 100 * (sp500_data['Adj Close_SPY'].shift(-50) - sp500_data['Adj Close_SPY']) / sp500_data['Adj Close_SPY']

# Drop NaN values that may result from the shift operation
ROI_365d = ROI_365d.dropna()

# Check if ROI_30d is empty
if ROI_365d.empty:
    print("No valid data for ROI calculation over 30 days. Ensure the dataset covers at least 30 days.")
else:
    # Plot the ROI
    ROI_365d.plot()
    plt.ylabel('% Return on Investment after 100 days')
    plt.title('S&P 500 ROI After 100 Days Based on Adj Close Price')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Select the column you want to apply the rolling window to, e.g., 'Adj Close_SPY'
column = 'Adj Close_SPY'
rolling = sp500_data[column].rolling(50, center=True)

# Create a DataFrame with the original data and the rolling mean and median
data = pd.DataFrame({
    'input': sp500_data[column],
    'one-year rolling_mean': rolling.mean(),
    'one-year rolling_median': rolling.median()
})

# Plot the data
ax = data.plot(style=['-', '--', ':'])
ax.lines[0].set_alpha(0.3)  # Adjust the transparency of the first line

# Show the plot
plt.show()
```
