```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Загружаем данные Iris
iris = load_iris()
X_iris = iris.data   # Признаки
y_iris = iris.target # Целевые значения

# Делим данные на обучающую и тестовую выборки
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)

print("Обучающая выборка:")
print(Xtrain[:5], ytrain[:5])  # Показываем первые 5 элементов
print("Тестовая выборка:")
print(Xtest[:5], ytest[:5])
```

```python
from sklearn.naive_bayes import GaussianNB  # правильное имя
model = GaussianNB()  # создаем модель
model.fit(Xtrain, ytrain)  # обучаем модель
y_model = model.predict(Xtest)  # предсказания на тестовых данных
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)
model.fit(X_iris)
X_2D = model.transform(X_iris)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
X_iris = iris.data  # Feature columns (Sepal and Petal dimensions)
```

```python
import pandas as pd
iris_df = pd.DataFrame(X_iris, columns=iris.feature_names)
iris_df['GMM_Cluster'] = y_gmm  # Add cluster labels to the DataFrame
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x=iris_df[iris.feature_names[0]], 
    y=iris_df[iris.feature_names[2]], 
    hue=iris_df['GMM_Cluster'], 
    palette='viridis'
)
plt.title("Gaussian Mixture Model Clustering")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[2])
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load iris data
iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['species'] = iris_data.target  # Add species column
```

```python
iris['PCA1'] = X_2D[:, 0]  # First PCA component
iris['PCA2'] = X_2D[:, 1]  # Second PCA component
iris['cluster'] = y_gmm    # Cluster labels from GMM
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.FacetGrid(iris, col="cluster", hue="species", height=4, aspect=1.2, palette="viridis")
g.map_dataframe(sns.scatterplot, x="PCA1", y="PCA2")
g.add_legend()

plt.show()
```

```python
sns.scatterplot(data=iris, x="PCA1", y="PCA2", hue="cluster", style="species", palette="viridis")
plt.title("Clusters and Species after GMM")
plt.show()
```

```python
from sklearn.manifold import Isomap
from sklearn.datasets import load_digits

# Load the digits dataset
digits = load_digits()

# Apply Isomap
iso = Isomap(n_components=2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)

# Output the shape of the reduced data
print(data_projected.shape)
```

```python
plt.scatter(data_projected[:, 0], data_projected[:,1], c=digits.target, edgecolor='none', alpha=0.5, cmap=plt.cm.get_cmap('viridis', 10))
plt.colorbar(label='digit label', ticks=range(10))
plt.clim(-0.5, 0.5)
```
