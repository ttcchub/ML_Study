```python
def f (x,y):
    return np.sin(x) ** 10 + np.cos(10+x+y)*np.cos(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a grid of x, y values
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) ** 10 + np.cos(10 + Y * X) * np.cos(X)

# Create a contour plot
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

# Display the image with color shading
plt.imshow(Z, extent=[0, 5, 0, 5], origin="lower", cmap='RdGy', alpha=0.5)

# Add a colorbar to show the color scale
plt.colorbar()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random data with a specific seed for reproducibility
rng = np.random.default_rng(1701)
data = rng.normal(size=1000)

# Plotting the histogram
plt.hist(data, bins=30, density=True, alpha=0.5,
         histtype='stepfilled', color='steelblue',
         edgecolor='none')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
rng = np.random.default_rng(1701)
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

# Define common histogram parameters
kwargs = {
    'bins': 30,
    'density': True,
    'alpha': 0.5,
    'histtype': 'stepfilled',
    'edgecolor': 'none'
}

# Plot histograms
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.show()
```

```python
from scipy.stats import gaussian_kde
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
rng = np.random.default_rng(1701)
x = rng.normal(size=1000)
y = rng.normal(size=1000)

# Calculate KDE
data = np.vstack([x, y])
kde = gaussian_kde(data)

# Define grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the density
plt.imshow(Z.reshape(Xgrid.shape),
           origin='lower', aspect='auto',
           extent=[-3.5, 3.5, -6, 6])
cb = plt.colorbar()
cb.set_label("density")
plt.show()
```
