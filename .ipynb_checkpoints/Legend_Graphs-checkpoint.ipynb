```python
import matplotlib.pyplot as plt

# Set parameters similar to 'seaborn-whitegrid' without directly using it
plt.rcParams.update({
    'grid.color': 'gray',
    'grid.linestyle': '-',
    'grid.linewidth': 0.5,
    'axes.edgecolor': 'black',
    'axes.facecolor': 'white',
    'axes.grid': True,
    'xtick.bottom': True,
    'ytick.left': True,
})

# Sample plot
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [1, 4, 9])
plt.show()
```

```python
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,10,1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sin' )
ax.plot(x, np.cos(x), '--r', label='Cosine' )
ax.axis('equal')
leg = ax.legend()
```

```python
ax.legend(loc='upper left', frameon='True')
fig
```

```python
ax.legend(loc='lower center', ncol=2)
fig
```

```python
ax.legend(frameon=True, fancybox=True, framealpha=1,
         shadow=True, borderpad=1);
fig
```

```python
y = np.sin(x[:,np.newaxis]+np.pi * np.arange(0,2,0.5))
lines = plt.plot(x,y)

plt.legend(lines[:2], ['first', 'second'], frameon=True);
```

```python

plt.plot(x, y[:,0],label='fisrt')
plt.plot(x, y[:, 1],label='second')
plt.plot(x, y[:,2:])
plt.legend(frameon=True);
```
