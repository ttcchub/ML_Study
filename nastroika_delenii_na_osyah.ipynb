```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)  # Corrected np*pi to np.pi
ax.plot(x, np.sin(x), lw=3, label='Sine')  # Sine plot
ax.plot(x, np.cos(x), lw=3, label='Cosine')  # Corrected np.con to np.cos

# Configure grid
ax.grid(True)  # Enable grid
ax.legend()  # Add legend

plt.show()
```

```python
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
fig
```
