```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic.groupby(['sex', 'class'], observed=True)['survived'].aggregate('mean').unstack()
```

```python
titanic.groupby(['sex', 'class'], observed=True)['survived'].aggregate('mean').unstack()
```

```python
titanic.pivot_table('survived', index='sex', columns='class', aggfunc='mean', observed=False)
```

```python
age = pd.cut(titanic['age'], [0,18,80])
titanic.pivot_table('survived', ['sex', age], 'class',observed=False)
```

```python
fare = pd.qcut(titanic['fare'],2)
titanic.pivot_table('survived', ['sex', age], [fare,'class'],observed=False)
```

```python
import pandas as pd

# Assuming 'data' is a dictionary
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Values': [10, 15, 10, 20, 15, 25],
    'Type': ['X', 'X', 'Y', 'Y', 'X', 'Y']
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Now apply pivot_table on the DataFrame
pivot_df = pd.pivot_table(df, values='Values', index='Category', columns='Type', aggfunc='mean')

# Print the resulting pivot table
print(pivot_df)
```
