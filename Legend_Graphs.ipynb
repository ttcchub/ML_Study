```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
cities = pd.read_csv('data/california_cities.csv')

# Correct variable assignments
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

# Create the scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(lon, lat, label=None,
            c=np.log10(population), cmap='viridis',
            s=area, linewidth=0, alpha=0.5)
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# Add legend for city area
for area_size in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area_size,
                label=f'{area_size} km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Area')

# Add title
plt.title('California Cities: Area and Population')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.legend import Legend  # Import the correct Legend class

# Create the figure and axis
fig, ax = plt.subplots()

# Initialize lines and styles
lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

# Plot the lines
for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2), styles[i], color='black')
    ax.axis('equal')

# Add the first legend
ax.legend(lines[:2], ['line A', 'line B'], loc='upper right')

# Add the second legend
leg = Legend(ax, lines[2:], ['line C', 'line D'], loc='lower right')
ax.add_artist(leg)  # Add the second legend to the plot

# Show the plot
plt.show()
```
